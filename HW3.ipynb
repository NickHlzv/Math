### 2
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.<br>

Пусть точки M1(x1, y1) и M2(x2, y2) посредством ортогонального преобразования переводятся соответствено в точки M'1 (x'1, y'1) и M'2(x'2, y'2). Требуется доказать, что отрезки M1M2 и M'1M'2 имеют равные длины.<br>

Доказательство:<br>

|M'1M'2|^2 = [x'2-x'1]^2 + [y'2-y'1]^2 + [a11(x2-x1) + a12(y2-y1)]^2 + [a21(x2-x1) + a22(y2-y1)]^2 =<br>

= (a11^2 + a21^2) (x2-x1)^2 + (a12^2 + a22^2) (y2-y1)^2 + 2 (a11a12 + a21a22)(x2-x1)(y2-y1) =<br>

= (x2-x1)^2 + (y2-y1)^2 = |M1M2|^2<br>

Итак, |M1M2| = |M'1M'2|

```python
#4.1
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5)
plt.grid(True)
plt.axis('scaled')
plt.show()

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))

print (x1, y1)
print (x2, y2)
```

```python
#4.2
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)

```
