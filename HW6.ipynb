### 7
Полученные кубики после разрубания, можно разбить на 4 категории<br>
8 кубиков с 3мя белыми гранями, 12 кубиков с 2мя белыми гранями, 6 кубиков с одной гранью, 1 полностью черный<br>
Посчитаем число комбинаций, которые можно получить вращая маленький кубик. Существует 6 вариантов зафиксировать верхнюю грань и 24 варианта расположения боковых граней при каждой фиксации. Таким образом, существует 24 способа повернуть маленький кубик.<br>
Для каждой группы кубиков количество подходящих вариантов равно количеству перестановок внутри группы умножить на количество подходящих поворотов:<br>
8!x3^8 — кубик можно вращать вокруг диагонали<br>
12!x2^12 — два варианта расположения белых граней<br>
6!x4^6 — при фиксации белой грани существует 4 варианта расположения<br>
1!x24 — для черного кубика подходят любые вращения<br>
Всего возможных комбинаций расположений кубиков = 27!x24^27<br>

Посчитаем вероятность в коде ниже

```python
(factorial(8)*(3**8)*factorial(12)*(2**12)*factorial(6)*(4**6)*24)/(factorial(27)*(24**27))
```
