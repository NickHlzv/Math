```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

### 1

```python
a1 = np.array([[1, 2, 3], [4, 0, 6], [7, 8, 9]])
b1 = np.array([12, 2, 1])
np.linalg.solve(a1, b1)
```

### 2

```python
#    -
x = np.linspace(-2, 2, 301)
y = x
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, x+2*y-1, label='1')
ax.plot(x, (3*x-7)/4, label='2')
ax.plot(x, y, (5*y-8*x+12)/2,  label='3')
ax.plot(x, y, (2*x-7)/5,  label='4')
ax.plot(x, y, (11*x+4*y-15)/7,  label='5')
```

```python
a2 = np.array([[1, 2, -1], [3, -4, 0], [8, -5, 2], [2, 0, -5], [11, 4, -7]])
b2 = np.array([1, 7, 12, 7, 15])
np.linalg.lstsq(a2, b2)
```

```python
#   
a5 = np.array([[1, 2, -1], [8, -5, 2]])
b5 = np.array([1, 12])
np.linalg.lstsq(a5, b5, rcond=None)
```

```python
#   
def Q(x, y, z):
    return(x**2+y**2+z**2)

x = np.linspace(-2, 2, 301)
y = x

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, Q(x,y, x+2*y-1), label='1')
ax.plot(x, y, Q(x, y, (5*y-8*x+12)/2),  label='3')
```
